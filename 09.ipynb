```python
from PIL import Image
import matplotlib.pyplot as plt
import numpy as np
```

```python
path = 'C:/Users/Luan/ThiGiacMayTinhSo7/img/00304.jpg'
# c nh
img = Image.open(path)
plt.imshow(img)
```

```python
# nh xm 
gray = img.convert("L")
```

```python
plt.imshow(gray, cmap='gray')
```

```python
array= np.array(gray)
print(array)
```

```python
# To mt hnh v mi
plt.figure()
plt.hist(array.flatten(), bins=128)
plt.title('Biu  Histogram ca nh')
plt.xlabel('Gi tr ca pixel')
plt.ylabel('S lng')
plt.show()
```

```python
# Lm ni bt ng bin
plt.figure()
plt.gray()
plt.contour(gray, origin="image")
plt.show()
```
