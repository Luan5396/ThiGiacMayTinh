```python
from PIL import Image
import matplotlib.pyplot as plt
import numpy as np
```

```python
path = 'C:/Users/Luan/ThiGiacMayTinhSo7/img/00304.jpg'
# đọc ảnh
img = Image.open(path)
plt.imshow(img)
```

```python
# ảnh xám 
gray = img.convert("L")
```

```python
plt.imshow(gray, cmap='gray')
```

```python
array= np.array(gray)
print(array)
```

```python
# Tạo một hình vẽ mới
plt.figure()
plt.hist(array.flatten(), bins=128)
plt.title('Biểu đồ Histogram của ảnh')
plt.xlabel('Giá trị của pixel')
plt.ylabel('Số lượng')
plt.show()
```

```python
# Làm nổi bật đường biên
plt.figure()
plt.gray()
plt.contour(gray, origin="image")
plt.show()
```

```python
path = 'C:/Users/Luan/ThiGiacMayTinhSo7/img/00304.jpg'
#Đọc ảnh và hiển thị ảnh với trục tọa độ
img = Image.open(path).convert("L")
# Hiển thị ảnh trên trục tọa độ
plt.imshow(img)
# Tạo một ình vẽ mới
plt.figure()
# Đặt màu sắc là xám
plt.gray()
# Vẽ biểu đồ đường biên (contour) của ảnh
plt.contour(img, origin="image")
# Đảm bảo tỷ lệ trục x và trục y trên biểu đồ là bằng nhau
plt.axis('equal')
# hiển thị ảnh
plt.show()
```
